```python
class Student():
    """Student class containing a student's name, GPA, and age.

    Overrides lt, eq, and hash dunders/magic functions.
    """
    def __init__(self, name, gpa, age):
        self.name = name
        self.gpa = gpa
        self.age = age

    def __lt__(self, other):
        """Evaluates whether this student is less than another student by
        comparing gpa, name, then age.

        Args:
            other (Student): the other student to compare

        Returns:
            bool: True if this student's gpa, name, or age (in that order) is
                less than the other student's.

        """
        return (self.gpa, self.name, self.age) < \
                (other.gpa, other.name, other.age)

    def __eq__(self, other):
        """Evaluates whether this student's name, gpa, and age are equal to
        the other student's name, gpa, and age.

        Args:
            other (Student): the other student to compare

        Returns:
            bool: True if this student's gpa, name, and age is equal to the
                other student's

        """
        return (self.name, self.gpa, self.age) == \
                (other.name, other.gpa, other.age) and \
                type(self) == type(other)

    def __hash__(self):
        """Computes a hash of a student's name and age.

        Returns:
            int: a hash of the student's name and age

        """
        return hash((self.name, self.age))

    def __str__(self):
        return '{0: <20}{1:4}{2:4}'.format(self.name, str(self.age), str(self.gpa))
```
